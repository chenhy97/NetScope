```python
def get_flow_latency_webget(n=100):
    import pandas as pd
    import pickle
    all_flows_data_file = "webget_2016.csv"
    flow_data = pd.read_csv(all_flows_data_file).groupby(["src_ip","dst_ip", "sport","dport"])["latency"].agg(list)
    sent_flow_data_file = "sent_flows_webget_2016.pickle"
    flows = pickle.load(open(sent_flow_data_file,"rb"))
    flow_latency = {}
    for key, value_list in flow_data.items():
        if tuple(key) in flows.keys():
            flow_latency[tuple(key)] = value_list[:n]
    return flow_latency
```

```python
flow_latency = get_flow_latency_webget(100)
```

```python
for flow, flow_lat in flow_latency.items():
    max_heap_max = 0
    max_heap_min = -1
    max_heap_count = 0
    max_heap_sum = 0
    n = 0
    p = 0.9
    for lat in (flow_lat):
        n += 1
        if max_heap_count < n * (1-p):
            max_heap_sum += lat
            max_heap_count += 1
            if lat < max_heap_min or max_heap_min == -1:
                max_heap_min = lat
            if lat > max_heap_max:
                max_heap_max = lat
        else:
            if lat > max_heap_max:
                max_heap_max = lat
                max_heap_min = max_heap_min + (max_heap_max - max_heap_min) >> 3
            if lat < max_heap_min or max_heap_min == -1:
                print(("{},{},[{},{}], {}, {}, {}".format(n, lat, max_heap_min, max_heap_max, max_heap_count, n * (1-p), max_heap_sum)))
                continue
            if max_heap_min< lat and lat < max_heap_max:
                max_heap_min = max_heap_min + (max_heap_max - max_heap_min) >> 3
        print(("{},{},[{},{}], {}, {}, {}".format(n, lat, max_heap_min, max_heap_max, max_heap_count, n * (1-p), max_heap_sum)))
    sorted_latency = sorted(flow_lat)
    print(sorted_latency[int(len(sorted_latency) * 0.9)], max_heap_min)
    break
```
