```python
COUNT_SKETCH_CELL_BIT_WIDTH = 16
TS_SKETCH_CELL_BIT_WIDTH = 48
```

```python
def total_mem(SKETCH_BUCKET_LENGTH=16):
    print(SKETCH_BUCKET_LENGTH )
    prev_ts_mem = TS_SKETCH_CELL_BIT_WIDTH * SKETCH_BUCKET_LENGTH * 3
    max_gap_mem = TS_SKETCH_CELL_BIT_WIDTH * SKETCH_BUCKET_LENGTH * 3
    min_gap_mem = TS_SKETCH_CELL_BIT_WIDTH * SKETCH_BUCKET_LENGTH * 3
    quantile_mem = TS_SKETCH_CELL_BIT_WIDTH * SKETCH_BUCKET_LENGTH * 3
    count_sketch_mem = COUNT_SKETCH_CELL_BIT_WIDTH * SKETCH_BUCKET_LENGTH * 3
    c_plus_mem = COUNT_SKETCH_CELL_BIT_WIDTH * SKETCH_BUCKET_LENGTH * 3
    c_minus_mem = COUNT_SKETCH_CELL_BIT_WIDTH * SKETCH_BUCKET_LENGTH * 3
    total_mem = prev_ts_mem + max_gap_mem + min_gap_mem + count_sketch_mem + c_plus_mem + c_minus_mem + quantile_mem
    return total_mem/8/1024
```

```python
phi = 0.8
perc_range = {}
for i in range(1, 65536):
    perc_res = round(i-i*phi)
    if perc_res not in perc_range.keys():
        perc_range[perc_res] = []
    perc_range[perc_res].append(i)
```

```python
command_str = "table_add percentile_match set_percentile_result {}->{} => {} 0"
for key, range in perc_range.items():
    start = range[0]
    end = range[-1]
    idx = key
    print(command_str.format(start, end, idx))
```
